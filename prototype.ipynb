```python
import requests
import logging
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
flipkart_url = "https://www.flipkart.com/search?q="+"iphone12"
```

```python
flipkart_url
```

```python
url_client = urlopen(flipkart_url)
```

```python
flipkart_page = url_client.read()
```

```python
flipkart_page
```

```python
# now we will beautify it using bs so that it looks neater

flipkart_html = bs(flipkart_page,'html.parser')
```

```python
# we fild div tag and class in which these link to be clicked are available
bigbox = flipkart_html.find_all("div",{"class":"_1AtVbE col-12-12"})
```

```python
len(bigbox)
```

```python
#we will get inside div tags to reach href

productlink = "https://www.flipkart.com"+bigbox[3].div.div.div.a['href']
```

```python
productlink
```

```python
# now we will read this product link to reach reviews


product_req = requests.get(productlink)
```

```python
product_req
```

```python
product_html = bs(product_req.text,'html.parser')
```

```python
comment_box = product_html.find_all("div",{"class":"_16PBlm"})
```

```python
len(comment_box)
```

```python
# to get the name from revies section
comment_box[0].div.div.find_all('p',{"class":"_2sc7ZR _2V5EHH"})[0].text
```

```python
# again we can put a for loop to get all the names of reviewers

for i in comment_box:
    print(i.div.div.find_all('p',{"class":"_2sc7ZR _2V5EHH"})[0].text)
```

```python
# to get the rating

comment_box[0].div.div.div.find_all('div',{"class":"_3LWZlK _1BLPMq"})[0].text
```

```python
# same we can put a for loop for all rating
for i in comment_box:
    print(i.div.div.div.find_all('div',{"class":"_3LWZlK _1BLPMq"})[0].text)
```

```python
# as they are titles and all
del bigbox[0:3]
```
